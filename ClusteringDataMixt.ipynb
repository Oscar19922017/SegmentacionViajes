```python
df=pd.read_csv("https://raw.githubusercontent.com/SteffiPeTaffy/machineLearningAZ/master/Machine%20Learning%20A-Z%20Template%20Folder/Part%204%20-%20Clustering/Section%2025%20-%20Hierarchical%20Clustering/Mall_Customers.csv")
```

```python
df.head()
```

```python
df.columns= ['CustomerID',"Genre",	'age', 'annual_income', 'spending_score']
```

```python

df
```

```python
X=df[["Genre"	,"age",	"annual_income",	"spending_score"]]
```

```python
X
```

```python
distance_matrix=gower.gower_matrix(X)
```

```python
data=pd.DataFrame(distance_matrix)
```

```python
data
```

```python
from pycaret.clustering import *
```

```python
#!pip install pycaret[full]
```

```python
cluster1=setup(data,session_id=123,log_experiment=True,log_plots=True,experiment_name="segmentacion")
```

```python
models()
```

```python
kmeans = create_model('kmeans', num_clusters = 4)
```

```python
plot_model(kmeans, plot = 'elbow')
```

```python
plot_model(kmeans)
```

```python
plot_model(kmeans, plot = 'distance')
```

```python
plot_model(kmeans, plot = 'distribution')
```

```python
plot_model(kmeans, plot = 'silhouette')
```

```python
kmeans_results = assign_model(kmeans)
kmeans_results.head()
```

```python
df["Cluster"]=kmeans_results.Cluster.values
```

```python
df
```

```python
cluster=df["Cluster"].unique().tolist()
resultados= {}
for i in cluster:
    grupo=df[df["Cluster"]==i]
    resultados[i] = grupo
```
